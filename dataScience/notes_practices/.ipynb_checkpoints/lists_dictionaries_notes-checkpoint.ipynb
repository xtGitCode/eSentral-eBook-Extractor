# Lists

Lists is a 0 based index datatype meaning the index of the first element starts at 0.

To store multiple items in a single variable.

```python
# create list
name_list = ['Jason','Chennai','Han Kook','Ceasar']

# accessing list element - Chennai
print(name_list[1])

# using negative index to access element - Han Kook
print(name_list[-2])

# extract portion in a list - Jason, Chennai and Han Kook
print(name_list[0:3])
```

```python
# Working with operators in a list
age_list = [34, 15, 25, 22]
# addition of two lists
name_age_list = name_list + age_list
name_age_list
```

```python
# Nested List
list_nest = [[1,2,3],['a','b','c']]
print(list_nest)

# access elements from nested list
print(list_nest[0][2]) # first column for list, second for list element
```

```python
# Update list
name_list.append('Mei') # append to update ONE item
print(name_list)

name_list.extend(['Leaf','Jorgond','Chalf']) # extend to update MULTIPLE items
print(name_list)
```

```python
# Insert items in list
name_list.insert(5,'Siri') 
print(name_list)
```

```python
# Delete items in list
del name_list[6]
print(name_list)
```

```python
# Length of List
print(len(name_list))
```

```python
# Sort list
name_list.sort()
print(name_list)
```

# Dictionary
mutable datatypes -> can be updated after they are created

written in key:value pair format inside curly braces eg. {Key1:Value1, Key2:Value2...}

Key needs to be unique. There is no index as the elements are not ordered. The key is used to retrieve the value



```python
# Create Dictionary

dict_city = {1:'Kuala Lumpur',2:'Kelantan',3:'Kuantan'}
print(dict_city)
```

```python
# Create mix dictionary

dict_mix = {'name1':'Boris',133:322,'list':['city','road',89]}
print(dict_mix)
```

```python
# Access dictionary elements

dict_salesid = {
    'SID1': 'Fiat',
    'SID2': 'Mercedes',
    'SID3': 'Maruti',
    'SID4': 'Volkswagen',
    'SID5': 'Kia'
}
dict_salesinfo = {
    'SID': 'Fiat',
    'Sales': 20000,
    'LaunchDay': 'Wed',
    'Cost': 500000
}

# Length of dict
print(len(dict_salesid))
```

```python
# Extracting dictionary element using its key
# example 1
print(dict_salesinfo['SID'])

# example 2
salesID = 'SID2'
if salesID in dict_salesid:
    name = dict_salesid[salesID]
    print('Sales ID is {}, Sales name is {}'.format(salesID,name))
```

```python
# Retrieve using get function

LaunchDay = dict_salesinfo.get('LaunchDay')
LaunchDay
```

```python
# items() function
# item function converts a dictionary item into a tuple
dict_items = dict_salesinfo.items()
print(dict_items)
```

```python
# Convert list to dictionary object using dict(list)

list_salesinfo = [['SID','Fiat'],['Sales',20000]]
list2dict_salesinfo = dict(list_salesinfo)
print(list2dict_salesinfo)
```
